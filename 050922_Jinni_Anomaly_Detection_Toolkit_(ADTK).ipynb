```python
from pathlib import Path
#path=Path('./adtk/docs/notebooks/data/quickstart/')
path=Path('.')
path
```

```python
import pandas as pd
df = pd.read_csv(path/"test_task_for_data_analyst___data_sample_2022-08-18T21_20_07.621729Z.csv", index_col="hire_reported", parse_dates=True, squeeze=True)
df.index = pd.to_datetime(df.index, utc=True, infer_datetime_format=True)
df.sort_index(inplace=True)
df
```

```python
df_key = df[(df['candidates_city']=='Киев') & (df['hire_salary']>0) & (df['primary_keyword_candidate']=='Java')].copy()
df_key
```

```python
df_key.columns
```

```python
#
select = 'hire_salary'
ts = df_key[select].copy()
ts
from adtk.data import validate_series
ts = validate_series(ts)
print(ts)
```

```python
from adtk.detector import ThresholdAD
threshold_ad = ThresholdAD(high=5000, low=1000)
anomalies = threshold_ad.detect(ts)

from adtk.visualization import plot
plot(ts, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag="marker");
```

```python
from adtk.detector import QuantileAD
#quantile_ad = QuantileAD(high=0.975, low=0.025)
quantile_ad = QuantileAD(high=0.99, low=0.01)
anomalies = quantile_ad.fit_detect(ts)

plot(ts, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag="marker");
```

```python
anomalies[anomalies == True]
```

```python
from adtk.detector import InterQuartileRangeAD
iqr_ad = InterQuartileRangeAD(c=1.5)
anomalies = iqr_ad.fit_detect(ts)

plot(ts, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag="marker");
```

```python
anomalies[anomalies == True]
```

```python
select = ['experience_years', 'hire_salary']
df_t = df_key[select].copy()
df_t
```

```python
#
def myTransformationFunc(df):
    return (df['hire_salary'] / df['experience_years']).rename("salary/exp_years")
```

```python
from adtk.transformer import CustomizedTransformerHD
customized_transformer = CustomizedTransformerHD(transform_func=myTransformationFunc)
s = customized_transformer.transform(df_t)
plot(pd.concat([df_t, s], axis=1), ts_linewidth=1, curve_group=[("hire_salary", "experience_years"), "salary/exp_years"]);
```

```python
df_key['hire_salary'].median()
```

```python
s.median()
```

```python
from adtk.detector import VolatilityShiftAD
volatility_shift_ad = VolatilityShiftAD(c=0.5, side='positive', window=30)
anomalies = volatility_shift_ad.fit_detect(s)
plot(s, anomaly=anomalies, anomaly_color='red');
```
